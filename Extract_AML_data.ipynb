```python
import pandas as pd

# Define a chunk size
chunk_size = 10000
chunks = []

# Read the CSV file in chunks
for chunk in pd.read_csv('/kaggle/input/ibm-transactions-for-anti-money-laundering-aml/HI-Large_Trans.csv', chunksize=chunk_size):
    # Filter each chunk
    filtered_chunk = chunk[chunk['Is Laundering'] == 1]
    chunks.append(filtered_chunk)

# Concatenate all filtered chunks into a single DataFrame
df1_filtered = pd.concat(chunks, ignore_index=True)
```
